```python
 import matplotlib.pyplot as plt
def lms(t,x,n,mu):
    w = np.zeros(n)
    y = np.zeros(n)
    err = np.zeros(n)
    for i in range(499):
        y[i] = w[i]*x[i]
        err[i] = x[i] - y[i]
        w[i+1] = w[i] + mu*(err[i]*x[i])
    plt.rcParams["figure.figsize"] = (13,11)
    figure, axis = plt.subplots(3,2)
    axis[0, 0].plot(t, x)
    axis[0, 0].set_title("Orginal Signal")
    axis[0, 1].plot(t, y)
    axis[0, 1].set_title("Desired Signal")
    axis[1, 0].plot(t, err)
    axis[1, 0].set_title("Error Signal")
    axis[1, 1].plot(t, w)
    axis[1, 1].set_title("W signal")
    axis[2, 0].plot(t, x)
    axis[2, 0].plot(t, y)
    axis[2, 0].set_title("Orginal vs Desired Signal")
    axis[2, 1].plot(t, y)
    axis[2, 1].plot(t, err)
    axis[2, 1].set_title("Desired vs Error Signal")
```

```python
import numpy as np
t = np.ones(500)
for i in range(1,500):
    t[i] = t[i-1] + 1
n = 500    #defining the length of the time numpy array for t
import math
x = np.zeros(n)
for i in range(500):
    x[i] = x[i] + math.sin(t[i])
mu = 0.1  #defining the value for mu as 0.1 now we will later do it for 0.01
```

```python
lms(t,x,n,mu)
#for sine function with mu 0.1
```

```python
mu = 0.01
lms(t,x,n,mu)
#for sine function with mu 0.01
```

```python
import numpy as np
t1 = np.ones(500)
for i in range(1,500):
    t1[i] = t1[i-1] + 1
n1 = 500    #defining the length of the time numpy array for t
import math
x1 = np.zeros(n1)
for i in range(500):
    x1[i] = x1[i] + math.cos(t1[i])
mu = 0.1  #defining the value for mu as 0.1 now we will later do it for 0.01
```

```python
lms(t1,x1,n1,mu)
#for cos function with mu 0.1
```

```python
mu = 0.01
lms(t1,x1,n1,mu)
#for cos function with mu 0.01
```
